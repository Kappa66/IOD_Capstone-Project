## Convert Published Date to Date time object

```python
def normalize_to_ymd(s):
    """
    Take a string s and return 'YYYY-MM-DD':
    - 'YYYY'        -> 'YYYY-01-01'
    - 'YYYY-M' or 'YYYY-MM' -> 'YYYY-MM-01'
    - 'YYYY-M-D', 'YYYY-MM-D', etc -> zero-pad and keep it
    - else: return as-is (to let to_datetime coerce it or fail to NaT)
    """
    if pd.isna(s):
        return None
    s = str(s).strip()
    # year only
    if re.fullmatch(r'\d{4}', s):
        return f'{s}-01-01'
    # year + month (1 or 2 digits)
    m = re.fullmatch(r'(\d{4})-(\d{1,2})$', s)
    if m:
        year, month = m.groups()
        return f'{year}-{month.zfill(2)}-01'
    # year-month-day (any digit count) → zero-pad each part
    parts = s.split('-')
    if len(parts) == 3 and all(p.isdigit() for p in parts):
        y, m, d = parts
        return f'{y.zfill(4)}-{m.zfill(2)}-{d.zfill(2)}'
    # otherwise leave it and let to_datetime errors='coerce'
    return s
```

```python
# 1) normalize every string
books_clean['publishedDate_norm'] = books_clean['publishedDate'].apply(normalize_to_ymd)
```

```python
# 2) parse to datetime64[ns], coercing any bad ones to NaT
books_clean['publishedDate_dt'] = pd.to_datetime(
    books_clean['publishedDate_norm'],
    format='%Y-%m-%d',
    errors='coerce'
)
```
